```python
def MACD(data):
    k1= 2/(12+1)
    k2= 2/(26+1)
    for index in data[:-1].index:
        data.loc[index+1, 'EMA_1'] = k1*data.loc[index+1, '4. close'] + (1-k1)*data.loc[index, 'EMA_1']
#         data.loc[index+1, 'EMA_2'] = k2*data.loc[index+1, '4. close'] + (1-k2)*data.loc[index, 'EMA_2']
#     data['MACD'] = data['EMA_1'] - data['EMA_2']
    return data 
```

```python
def calculate_k(data):
    
    close = data['4. close']
    highest_hi = data['2. high'].rolling(window=10).max()
    lowest_lo = data['3. low'].rolling(window=10).min()
    data['per_k_stoch_10'] = (close- lowest_lo) / (highest_hi - lowest_lo) *100
    
    return data

def calculate_d(data):
    
    data['per_d_stoch_10'] = data['per_k_stoch_10'].rolling(window=10).mean()
    
    return data

def stochastic_oscillator(data):
    data = calculate_k(data)
    data= calculate_d(data)
    
    return data

    
    
    
```

```python
stochastic_oscillator(data_dated)
```

```python
def accumulation_distribution(data):
    ##data is ordered by date with the oldest having an index of 0
    data.reset_index(inplace = True) 
    values = pd.Series(index = data.index)
    first_idx = data.index.values[0]
    
    for idx in data.index.values:
        today = data.loc[idx]
        close, high, low, volume = today['4. close'], today['2. high'], today['3. low'], today['6. volume']
        
        CLV= ((close-low)- (high-close))/(high-low)
        
        values[idx] = values[idx-1] + CLV * volume if idx != first_idx else 0
        
    data['a/d']=values
    
    return data
```

```python
accumulation_distribution(data_dated)
```

```python
def obv(data):
    for index in data[:1].index:
        data.loc[index+1, 'OBV'] = abs(data.loc[index+1, '6. volume']- data.loc[index, '6. volume'])
    return data

    
```

```python
obv(data_dated)
```

```python
def BBANDS(data):
    data['TP']= (data['2. high'] + data['3. low'] + data['4. close'])/3
    data['MidBand']= data['TP'].rolling(window=20).mean()
    data['Std']= data['TP'].rolling(window=20).std()
    
    data['UpperBand'] = data['MidBand'] + (data['Std'] * 2)
    data['LowerBand'] = data['MidBand'] - (data['Std'] * 2)
    
    data = data.drop(['Std', 'TP'], axis=1)
    
    return data
```
