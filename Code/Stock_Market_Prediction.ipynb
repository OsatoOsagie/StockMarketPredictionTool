```python
from alpha_vantage.timeseries import TimeSeries
from pprint import pprint
from alpha_vantage.techindicators import TechIndicators
from datetime import date
from datetime import timedelta

import matplotlib.pyplot as plt 
import sys
import pandas as pd
import time

```

```python
# Obtaining stock data from alpha vantage 
sym='AAPL'

#obtains stock data from the previous day date to 2018-11-29
def stock_data(symbol):
    ts = TimeSeries(key='SUXOFAIGXM6HEP9Y', output_format='pandas')
    data, meta_data = ts.get_daily_adjusted(symbol=sym, outputsize='full')
    data_date_changed = data[:'2018-11-29']
    return data_date_changed

data_dated = stock_data(sym)  
```

```python
time.sleep(20)
```

```python
# *****************************Technical Indicators*******************************************************************
```

```python
# *****************************SMA*******************************************************************

# getting the date of the previous day
yday= date.today()- timedelta(1) 

# technical indicators from alphavantage
ti = TechIndicators(key = 'SUXOFAIGXM6HEP9Y', output_format = 'pandas')

# obtaining sma inidcator from alpha vantage
stockSMA, meta_data= ti.get_sma(symbol = sym, interval = 'daily',time_period = 20)

# resting index so that date is no longer the index
stockSMA.reset_index(inplace = True) 

# sorting date column in decending order
stockSMA.sort_values(by=['date'], inplace=True, ascending=False)

# filtering date 
mask = (stockSMA['date'] > '2018-11-29') & (stockSMA['date'] <=pd.to_datetime(yday))

stockSMA = stockSMA.loc[mask]


stockSMA.head(20)

# resetting index 
stockSMA.set_index('date', inplace=True)


data_dated['SMA']= stockSMA[['SMA']].copy()

    
```

```python
# *****************************EMA*******************************************************************


# obtaining EMA inidcator from alpha vantage
stockEMA,q= ti.get_ema(symbol = sym, interval = 'daily', time_period = 20)

# resting index so that date is no longer the index
stockEMA.reset_index(inplace = True) 

# sorting date column in decending order
stockEMA.sort_values(by=['date'], inplace=True, ascending=False)

# filtering date 
mask = (stockEMA['date'] > '2018-11-29') & (stockEMA['date'] <=pd.to_datetime(yday))

stockEMA = stockEMA.loc[mask]


stockEMA.head(20)

# resetting index 
stockEMA.set_index('date', inplace=True)


data_dated['EMA']= stockEMA[['EMA']].copy()

```

```python
# *****************************MACD*******************************************************************

# obtaining MACD inidcator from alpha vantage
stockMACD,q= ti.get_macd(symbol = sym, interval = 'daily', fastperiod = 12, slowperiod = 26, signalperiod = 9)

# resting index so that date is no longer the index
stockMACD.reset_index(inplace = True) 

# sorting date column in decending order
stockMACD.sort_values(by=['date'], inplace=True, ascending=False)


# filtering date 
mask = (stockMACD['date'] > '2018-11-29') & (stockMACD['date'] <=pd.to_datetime(yday))

stockMACD = stockMACD.loc[mask]


stockMACD.head(20)

# resetting index 
stockMACD.set_index('date', inplace=True)


data_dated['MACD']= stockMACD[['MACD']].copy()



```

```python
# ***********************************Bollinger Bands*******************************************************************
stockbbands, q = ti.get_bbands(symbol = sym, interval = 'daily', time_period = 20, nbdevup = 2, nbdevdn = 2, matype = 0)

# resting index so that date is no longer the index
stockbbands.reset_index(inplace = True) 

# sorting date column in decending order
stockbbands.sort_values(by=['date'], inplace=True, ascending=False)

# filtering date 
mask = (stockbbands['date'] > '2018-11-29') & (stockbbands['date'] <=pd.to_datetime(yday))

stockbbands = stockbbands.loc[mask]


stockbbands.head(20)

# resetting index 
stockbbands.set_index('date', inplace=True)


data_dated['Real Lower Band']= stockbbands[['Real Lower Band']].copy()
data_dated['Real Upper Band']= stockbbands[['Real Upper Band']].copy()
data_dated['Real Middle Band']= stockbbands[['Real Middle Band']].copy()
```

```python
# ***********************************Stochastic oscillator*******************************************************************
stockStoch, meta_data = ti.get_stoch(symbol=sym, interval='daily')

# resting index so that date is no longer the index
stockStoch.reset_index(inplace = True) 

# sorting date column in decending order
stockStoch.sort_values(by=['date'], inplace=True, ascending=False)


# filtering date 
mask = (stockStoch['date'] > '2018-11-29') & (stockStoch['date'] <=pd.to_datetime(yday))

stockStoch = stockStoch.loc[mask]


stockStoch.head(20)

# resetting index 
stockStoch.set_index('date', inplace=True)


data_dated['SlowK']= stockStoch[['SlowK']].copy()
data_dated['SlowD']= stockStoch[['SlowD']].copy()
```
