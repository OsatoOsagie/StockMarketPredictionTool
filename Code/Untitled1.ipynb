```python
from alpha_vantage.timeseries import TimeSeries
from pprint import pprint
from alpha_vantage.timeseries import TimeSeries 
import matplotlib.pyplot as plt 
import sys
```

```python
# Obtaining stock data from alpha vantage 
symbol='AAPL'

def stock_data(symbol):
    ts = TimeSeries(key='SUXOFAIGXM6HEP9Y', output_format='pandas')
    data, meta_data = ts.get_daily_adjusted(symbol='AAPL', outputsize='full')
    data_date_changed = data[:'2018-11-29']
    return data_date_changed

data_dated = stock_data(symbol)  
```

```python
data_dated['SMA'] = ta.SMA(data_dated['4. close'], timeperiod = 20)
```

```python
data_dated['EMA'] = ta.EMA(data_dated['4. close'], timeperiod = 20)
```

```python
def Stoch(close,high,low):
    slowk, slowd = ta.STOCH(high, low, close)
    stochSell = ((slowk < slowd) & (slowk.shift(1) > slowd.shift(1))) & (slowd > 80)
    stochBuy = ((slowk > slowd) & (slowk.shift(1) < slowd.shift(1))) & (slowd < 20)
    return stochSell,stochBuy, slowk,slowd
```
